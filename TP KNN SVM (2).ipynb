```python
# Nom : [raji]    Prenom : [ Ismail]    Master : GLCC
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

# ===== 1. Dataset =====
X = np.array([[1,2],[2,3],[3,3],[6,5],[7,7],[8,6]])
y = np.array([0,0,0,1,1,1])

# ===== 2. Modèles =====
knn = KNeighborsClassifier(n_neighbors=3).fit(X, y)  # KNN
svm = SVC(kernel='linear').fit(X, y)                # SVM

# ===== 3. Point à tester =====
p = np.array([[5,5]])

# ===== 4. Affichage des points =====
plt.scatter(*X[y==0].T, label="Classe 0")  # classe 0
plt.scatter(*X[y==1].T, label="Classe 1")  # classe 1
plt.scatter(*p.T, marker='x', s=100, label="Test")  # point test

# ===== 5. Frontière SVM =====
xx, yy = np.meshgrid(np.linspace(0,10,100), np.linspace(0,10,100))
Z = svm.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
plt.contour(xx, yy, Z, alpha=0.3)

# ===== 6. Résultat =====
print("KNN:", knn.predict(p))
print("SVM:", svm.predict(p))

plt.legend()
plt.title("KNN vs SVM")
plt.show()
```
